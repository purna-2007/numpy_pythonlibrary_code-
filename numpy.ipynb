```python
import numpy as np
a=np.array([[9,0,2],[6,3,4]])
b=np.array([[4,5,3],[8,6,9]])
result=np.concatenate((a,b))
print(result)
```

```python
import numpy as np
a=np.array([[9,0,2],[6,3,4]])
b=np.array([[4,5,3],[8,6,9]])
c=np.array([[8,4,6],[9,6,7]])
result=np.concatenate(((a.flatten(),b.flatten(),c.flatten())))
print(result)
```

ravel is a change an entire array also and flatten has only can change in change that only array

```python
import numpy as np
arr=np.array([[9,0,2],[6,3,4]])
res=arr.flatten()
res[2]=31
print(res)
print(arr)
```

```python
import numpy as np
arr=np.array([[9,0,2],[6,3,4]])
res=arr.ravel()
res[2]=31
print(res,'\n')
print(arr)
```
